EDGE DETECTION


```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np


image = cv2.imread("sample_data/messi.jpg")
#cv2_imshow(image)

canny_image = cv2.Canny(image,120,150)
#cv2_imshow(canny_image)

#EROSION--eroding the pixels at the boundary
kernel = np.ones((1,1),np.uint8)
erode_image = cv2.erode(canny_image,kernel,iterations=1)
#cv2_imshow(erode_image)

#DILATION--adding pixels to the photo
kernel = np.ones((1,1),np.uint8)
dilate_image = cv2.dilate(canny_image,kernel,iterations=1)
#cv2_imshow(dilate_image)

dis = np.hstack((canny_image,erode_image,dilate_image))
cv2_imshow(dis)

```

REMOVING THE NOISE FROM AN IMAGE


```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

img = cv2.imread("lion.jpg")
#cv2_imshow(img)

i1=cv2.fastNlMeansDenoisingColored(img,None,25,14,7)  #fucntion to remove the noise in image

dis = np.hstack((img,i1))
cv2_imshow(dis)
```
